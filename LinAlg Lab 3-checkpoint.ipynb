$$ A = C\cdot [3,2],  B = C\cdot [5,4] $$

$$A = C\cdot  \begin{bmatrix} 3\\2 \\\end{bmatrix} ,

$$
S = C_1 \cdot [2,4], C_2 \cdot [1,3] 
$$

$$S = \begin{Bmatrix} C_1 \cdot\begin{bmatrix} 2\\4 \\\end{bmatrix}, 
C_2 \cdot \begin{bmatrix} 1\\3 \\\end{bmatrix}\end{Bmatrix} $$

```python
vectA = np.array([2,4])
vectB = np.array([1,3])
R = np.arange(-10,10,1)

c1, c2 = np.meshgrid(R,R) #creates a range of 2D values of R and R
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
vectR = vectA + vectB
spanRx = c1*vectA[0] + c2*vectB[0]
spanRy = c1*vectA[1] + c2*vectB[1]
plt.scatter(R*vectA[0],R*vectA[1], color='red', label="Vector R")
plt.scatter(spanRx,spanRy, s=5, alpha=0.50, color='blue', label="SpanX,Y")

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend()
plt.grid()
plt.show()
```
